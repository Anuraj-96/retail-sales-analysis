Read Data From File

```python
import pandas as pd
df = pd.read_csv('orders.csv')
```

Rename column names in lower case and replacing space with underscore

```python
df.rename(columns=lambda x: x.lower().replace(' ', '_'), inplace=True)
```

Handle Null Values

```python
df.describe(include='all')
```

```python
df["ship_mode"].unique() # Need to replace 'Not Available', 'unknown' as nan
df['ship_mode'] = df['ship_mode'].replace(['Not Available', 'unknown'], pd.NA)

df['ship_mode'].fillna(df['ship_mode'].mode()[0], inplace=True) #Has 1 missing. fill with most frequent value
```

Convert order_date to Datetime


```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
```

Create new columns : discount , sale_price and profit

```python
df['discount_amount'] = (df['list_price'] * df['discount_percent'] ) / 100
df['sale_price'] = df['list_price'] - df['discount_amount']
df['profit'] = (df['sale_price'] - df['cost_price']) * df['quantity']
df['profit'] = df['profit'].round(2)
```

```python
df.to_csv('retail_data.csv', index=False)
files.download('retail_data.csv')
```
